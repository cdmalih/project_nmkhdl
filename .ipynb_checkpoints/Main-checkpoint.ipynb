```python
# df_copy["Addresses"][0].split(",")[-1]
arr_provinces = []

for i in range(len(df_copy["Addresses"])):
    if df_copy["Addresses"][i] != "NAN" and type(df_copy["Addresses"][i]) == str:
        if check_list_province(df_copy["Addresses"][i].split(",")[-1].strip(), Lisst_Province) == False:
            arr_provinces.append("NAN")
        else:
            arr_provinces.append(df_copy["Addresses"][i].split(",")[-1].strip())
    else:
            arr_provinces.append("NAN")
print(len(arr_provinces) , len(df_copy["Addresses"]))
# df_copy["Provinces"].unique()
```

### Modeling and Predict prices

```python
X = df_newcp.iloc[: , :-1]
y = df_newcp.iloc[: , -1]
```

#### train test plit

```python
### Ussing ScalerStandard for X data

scaler = StandardScaler()
X = scaler.fit_transform(X)
X
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size= 0.8 , random_state= 83)
print(f"Shape X_train = ({X_train.shape})" )
print(f"Shape X_test = ({X_test.shape})" )
print(f"Shape y_train = ({y_train.shape})" )
print(f"Shape y_test = ({y_test.shape})" )
```

```python
rdfr_model = RandomForestRegressor(max_depth= 5 , n_jobs=2 , random_state= 83 )
rdfr_model.fit(X_train , y_train)
```

```python
predict = rdfr_model.predict(X_test)
predict
```

```python
mean_squared_error(y_test , predict , squared=False)
```

```python
r2_score(y_test, predict)
```
