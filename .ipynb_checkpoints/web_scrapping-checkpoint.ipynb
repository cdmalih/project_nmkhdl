```python
### 
def scrap_one_link(link_news , i , Titles  , Users   , Link_users  , Times , Prices , Squares  , Number_badrooms , Number_wcrooms  ,Addresses , Direction_houses,  Direction_balconys ):
#     Titles = []
#     Users = []
#     Link_users = []
#     Times = []
#     Prices = []
#     Squares  = []
#     Number_badrooms = []
#     Number_wcrooms = []
#     Addresses = []
#     Direction_houses = []
#     Direction_balconys = []
#     Direction_houses,  Direction_balconys 

    for link_new in link_news:
        if link_new == "https://batdongsan.vn/-r264105" or link_new == "https://batdongsan.vn/-r261591" or link_new == "https://batdongsan.vn/-r261964" or link_new == "https://batdongsan.vn/-r261899" or link_new == "https://batdongsan.vn/du-an-nha-o-lien-ke-ngay-trung-tam-thanh-pho-r250250":
            break
        html_text = requests.get(link_new).text
        soup = bs(html_text , "html.parser")

        ## Get title of the news
        title_soup = soup.find("h1" , class_ = "uk-panel-title")
        try:
            title = title_soup.span.get_text().strip()
        except:
            title = "NAN"
        if title == "NAN":
            break
        Titles.append(title)

        # Get Name of their own's new
        user_name = soup.find("div" , class_ = "name")
        user = user_name.get_text().strip()
        Users.append(user)

#         Get the User's Link 
        link_user = user_name.a["href"]
        Link_users.append(link_user)

#         Get the time was posted
        time = soup.find("time" , class_ = "timeago")
        time_post = time['datetime']
        date_format = "%Y-%m-%d %H:%M:%S"
        Times.append(datetime.datetime.strptime(time_post , date_format))

            # Get price of their own's new
        soup_price = soup.find("strong" , class_ = "price")
        price = " ".join(soup_price.get_text().strip().split(" "))

        if price != 'Thỏa thuận':
            price_re = re.match(r'^([\d.]+)\s*(\D+)$',price)
            unit = price_re.group(2) 
            if unit == 'triệu':
                price=  float(price_re.group(1)) / 1000
            else:
                if unit == 'tỷ':
                    price =  price_re.group(1)
        else:
            matches = re.findall(r'(\d+(\.\d+)?)\s*(tỷ|tỉ|ty)(\d{0,3})?', title, re.IGNORECASE)
            if matches:
                prices = [match[0] for match in matches]
                price = prices[-1]
        Prices.append(price)

    #     print(price)

            ### Get Square of a house.
                ## Get number of floor , number of bed_room , number of wc , address , ....


        soup_square = soup.find_all("div" , class_ = "param")
        for x in soup_square:
            ul_xs = x.ul
            for ul_x in ul_xs:
                check_list_info = ul_x.get_text().strip().split(":")
                if check_list_info != '':

                    if check_list_info[0] == "Diện tích":
                        try:
                            square = check_list_info[-1]
                            Squares.append(square)
                        except:
                            square = "NAN"
                            Squares.append(square)
                    if check_list_info[0] == "Hướng ban công":
                        try:
                            direct_bal = check_list_info[-1]
                            Direction_balconys.append(direct_bal)
                        except:
                            direct_bal = "NAN"
                            Direction_balconys.append(direct_bal)

                    if check_list_info[0] == "Hướng nhà":
                        try:
                            direct_house = check_list_info[-1]
                            Direction_houses.append(direct_house)
                        except:
                            direct_house = "NAN"
                            Direction_houses.append(direct_house)
                            
                    if check_list_info[0] == "Phòng ngủ":
                        try:
                            num_bed = check_list_info[-1]
                            Number_badrooms.append(num_bed)
                        except:
                            num_bed = "NAN"
                            Number_badrooms.append(num_bed)
                    if check_list_info[0] == "Phòng WC":
                        try:
                            num_wc = check_list_info[-1]
                            Number_wcrooms.append(num_wc)
                        except:
                            num_wc = "NAN"
                            Number_wcrooms.append(num_wc)
                    if check_list_info[0] == "Địa chỉ":
                        try:
                            address = check_list_info[-1]
                            Addresses.append(address)
                        except:
                            address = "NAN"
                            Addresses.append(address)
        if len(Squares) < len(Titles):
                square = "NAN"
                Squares.append(square)   
        if len(Number_wcrooms) < len(Titles):
                num_wc = "NAN"
                Number_wcrooms.append(num_wc)
        if len(Number_badrooms) < len(Titles):
                num_bed = "NAN"
                Number_badrooms.append(num_bed)
        if len(Addresses) < len(Titles):
                address = "NAN"
                Addresses.append(address)
        if len(Direction_houses) < len(Titles):
                direct_house = "NAN"
                Direction_houses.append(direct_house)
        if len(Direction_balconys) < len(Titles):
                direct_bal = "NAN"
                Direction_balconys.append(direct_bal)


        print(i)
        i+= 1
        
    return     i , Titles  , Users   , Link_users  , Times , Prices , Squares  , Number_badrooms , Number_wcrooms ,Addresses , Direction_houses,  Direction_balconys 
    
# Users
# link_users
# Times    
# Titles
# Prices
    
```

```python
#### Check each of link of each of page
def list_link_one_page(soup , max_each_page = 20):
    link_news = []
    object_news = soup.find_all("div" , class_ = "name")
    i = 1
    for object_new in object_news:
        if i > max_each_page:
            break
        link_news.append(object_new.a['href'])
        i += 1
    
    return link_news
# len(link_news)
```

```python
idx_value = 0
Titles = []
Users = []
Link_users = []
Times = []
Prices = []
Squares  = []
Number_badrooms = []
Number_wcrooms = []
Number_floor = []
Addresses = []
Direction_houses = []
Direction_balconys = []

for total_path in total_paths:
    html_total_text = requests.get(total_path).text
    soup = bs(html_total_text , "html.parser")
    link_list_news_of_onepage = list_link_one_page(soup)
    idx_value , Titles  , Users   , Link_users  , Times , Prices , Squares  , Number_badrooms , Number_wcrooms ,Addresses , Direction_houses,  Direction_balconys = scrap_one_link(link_list_news_of_onepage , idx_value , Titles  , Users   , Link_users  , Times , Prices , Squares  , Number_badrooms , Number_wcrooms ,Addresses , Direction_houses,  Direction_balconys )
    
```

```python
def loop_test(idx_value , total_paths , start , end , Titles  , Users   , Link_users  , Times , Prices , Squares  , Number_badrooms , Number_wcrooms ,Addresses , Direction_houses,  Direction_balconys):
    for total_path in total_paths[start : end] :
        html_total_text = requests.get(total_path).text
        soup = bs(html_total_text , "html.parser")
        link_list_news_of_onepage = list_link_one_page(soup)
        idx_value , Titles  , Users   , Link_users  , Times , Prices , Squares  , Number_badrooms , Number_wcrooms ,Addresses , Direction_houses,  Direction_balconys = scrap_one_link(link_list_news_of_onepage , idx_value , Titles  , Users   , Link_users  , Times , Prices , Squares  , Number_badrooms , Number_wcrooms ,Addresses , Direction_houses,  Direction_balconys )
    
```

```python
import threading
idx_value = 0
Titles = []
Users = []
Link_users = []
Times = []
Prices = []
Squares  = []
Number_badrooms = []
Number_wcrooms = []
Number_floor = []
Addresses = []
Direction_houses = []
Direction_balconys = []


# create 4 threads and start them
idx_value = [0 , 2500 , 5000 , 7500 ]
start = [0  , 125 , 250 , 375]
end = [125 , 250 , 375 , max_number_page]
threads = []
for i in range(4):
    thread = threading.Thread(target=loop_test , args=(idx_value[i] , total_paths , start[i] , end[i] ,Titles  , Users   , Link_users  , Times , Prices , Squares  , Number_badrooms , Number_wcrooms ,Addresses , Direction_houses,  Direction_balconys))
    thread.start()
    threads.append(thread)

# wait for all threads to finish
for thread in threads:
    thread.join()

print("Scraping finished")
```
